# EDA (EXPLORATORY DATA ANALYSIS)

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  2. Load Games Dataset

```python
file=pd.read_csv('game1.csv')
file.head(10)
```

## 3. To Know The Values Are Null Or Not

```python
file.isnull()
```

## 4. To Know How Many Values Are Null Values In Each Columns 

```python
count_of_missing_values=file.isnull().sum()
count_of_missing_values[0:12]
```

## 5. To Know What Datatype It Is

```python
file.turns.dtype
```

```python
file.increment_code	.dtype
```

```python
file.opening_ply.dtype
```

## 6. To Fill NaN Values With Some Other Value 

```python
file.fillna('not found')
```

##  7. To replace anything in columns

```python
file.winner.replace('black','brown')
```

## 8. Line Plot

```python
plt.figure(figsize=(14,6))
plt.title("games data line plot")
sns.lineplot(data=file,x='turns',y='opening_ply')
plt.show()
```

## 9. Bar Plot

```python
sns.barplot(data=file,x="winner",y="turns")
```

## 10. Heatmap

```python
plt.figure(figsize=(14,6))
sns.heatmap(file.isnull())
```

## 11. ***Distplot***

```python
sns.distplot(file['turns'].dropna(),color='blue',bins=10)
```

## 12. Boxplot

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='rated',y='opening_ply',data=file,palette='Dark2_r')
```

## 13. Scatterplot

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x='winner',y='turns',data=file,palette='Dark2_r')
```

## 14. Pairplot

```python
plt.figure(figsize=(14,10))
sns.pairplot(file)
```

## 15. Cleaning Dataset

```python
file.drop('opening_name',axis=1,inplace=True)
```
